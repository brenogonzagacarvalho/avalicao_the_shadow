## Análise de Comunicação

### Subtask:
Definir e aplicar métodos para identificar padrões e métricas relacionadas à comunicação dentro dos dados, focando na análise de texto das avaliações.


**Reasoning**:
Following the instructions, I will now perform text preprocessing on the 'Review' column. This involves converting all text to lowercase, removing punctuation, numbers, and special characters, and finally removing stopwords to clean the text for further analysis.



```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation, numbers, and special characters
    words = text.split()
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return ' '.join(words)

df['processed_review'] = df['Review'].apply(preprocess_text)

print("Text preprocessing completed. New 'processed_review' column created.")
print(f"First 5 processed reviews:\n{df['processed_review'].head().to_markdown(numalign='left', stralign='left')}")
```

**Reasoning**:
Following the instructions, I will now tokenize the pre-processed text in the 'processed_review' column. This involves splitting each review into individual words and storing them as a list of tokens in a new column.



```python
df['tokens'] = df['processed_review'].apply(lambda x: x.split())

print("Text tokenization completed. New 'tokens' column created.")
print(f"First 5 tokenized reviews:\n{df['tokens'].head().to_markdown(numalign='left', stralign='left')}")
```

## Análise de Colaboração

### Subtask:
Realizar análise de sentimento nas avaliações processadas para identificar padrões de apreciação ou crítica, indicando formas de 'colaboração' na formação de opiniões.


**Reasoning**:
As per the instructions, I will start by importing the necessary `SentimentIntensityAnalyzer` from `nltk.sentiment.vader`, download the 'vader_lexicon' if not already present, and then instantiate the analyzer. This prepares the environment for sentiment analysis.



```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon', quiet=True)
sia = SentimentIntensityAnalyzer()

print("SentimentIntensityAnalyzer initialized and vader_lexicon downloaded.")
```

**Reasoning**:
Following the instructions, I will apply the initialized `SentimentIntensityAnalyzer` to the `processed_review` column to calculate the 'compound' sentiment score for each review and store it in a new `sentiment_score` column.



```python
df['sentiment_score'] = df['processed_review'].apply(lambda review: sia.polarity_scores(review)['compound'])

print("New 'sentiment_score' column created with compound sentiment scores.")
print(f"First 5 rows with sentiment scores:\n{df[['processed_review', 'sentiment_score']].head().to_markdown(index=False, numalign='left', stralign='left')}")
```

**Reasoning**:
As instructed, I will now display the descriptive statistics for the newly created `sentiment_score` column to understand the distribution of sentiment scores. Then, I will calculate and print the average sentiment scores for reviews marked as 'Recomendado' (True) and 'Not Recomendado' (False).



```python
print("\nDescritive statistics for sentiment_score column:\n")
print(df['sentiment_score'].describe().to_markdown(numalign='left', stralign='left'))

print("\nAverage sentiment score for recommended reviews:\n")
print(df.groupby('Recomendado')['sentiment_score'].mean().to_markdown(numalign='left', stralign='left'))
```
